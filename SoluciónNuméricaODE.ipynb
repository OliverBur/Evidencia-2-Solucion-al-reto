```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0.0     # Tiempo de inicio
tf = 30.0     # Tiempo final
h = 0.01     # Paso
x0 = 20     # Condición inicial
y0 = 20     # Condición inicial
a = 1
b = 0.02
c = 1
d = 0.01
e = 0.5
```

```python
nt = int((tf-t0)/h) #número de pasos
x = np.empty(nt+1)
x[0]=x0
y = np.empty(nt+1)
y[0]=y0

for step in range(nt):
  Xp = a*x[step] - b*x[step]*y[step] - e*x[step]
  Yp = -c*y[step] + d*x[step]*y[step]-  e*y[step]
  
  Xpp = a*Xp - b*(Xp*y[step] + Yp*x[step]) + e*Xp
  Ypp = -c*Yp + d*(Xp*y[step] + Yp*x[step]) + e*Yp
  
  x[step+1] = x[step]+Xp*h+(1/2)*Xpp*(h**2)
  y[step+1] = y[step]+Yp*h+(1/2)*Ypp*(h**2)
```

Los valores de la función aproximada son:
