```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw
from scipy import stats
from scipy.stats import expon
```

```python
tiempo_lim = 30 ##límite de tiempo
presas = []  #array para evolución temporal número de presas
depred = [] #array para evolución temporal número de depredadores
pp =50.0  #número inicial de presas
dd =100.0 #número inicial de depredadores

##constantes para calcular los rates
c1=1.0
c2=0.005
c3=0.6

currTime=0 #tiempo actual
```

```python
dd_array = []
pp_array = []

while currTime < tiempo_lim:
  if (pp > 0 and dd > 0):
    rate1 = c1*pp 
    rate2 = c2*pp*dd
    rate3 = c3*dd

    rate = rate1+rate2+rate3

    p1 = rate1/rate
    p2 = rate2/rate
    p3 = rate3/rate

    c = np.random.multinomial(1,[p1,p2,p3])
    if c[0] == 1:
      pp +=1

    elif c[1] == 1:
      pp -= 1
      dd += 1

    elif c[2] == 1:
      dd -= 1

  elif (pp == 0 and dd == 0):
    currTime = tiempo_lim

  elif (pp>0 and dd==0):
    rate = c1*pp
    pp+=1

  elif (dd>0 and pp==0):
    rate = c3*dd
    dd -= 1
  currTime += np.random.exponential(1/rate)
  dd_array.append(dd)
  pp_array.append(pp)
```

```python

plt.plot(dd_array)
plt.plot(pp_array)
```
